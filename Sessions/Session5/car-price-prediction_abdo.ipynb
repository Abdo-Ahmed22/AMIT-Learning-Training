## **Data Preprocessing**

#### **Drop Unneeded Columns**

```python
df.head()
```

#### **Define Features X and Tanget y**

```python
X = df.drop(columns=['Price'])
y = df['Price']
```

```python
X
```

```python
y
```

#### **Data Encoding**

```python
X.head()
```

```python
encoder = LabelEncoder()

X['Brand'] = encoder.fit_transform(X['Brand'])
```

```python
X.head()
```

```python
X['Fuel Type'] = encoder.fit_transform(df['Fuel Type'])
X['Transmission'] = encoder.fit_transform(df['Transmission'])
X['Condition'] = encoder.fit_transform(df['Condition'])
X['Model'] = encoder.fit_transform(df['Model'])
```

```python
X
```

#### **Data Scaling**

```python
X.shape
```

```python
X[['Mileage']].shape
```

```python
scaler = MinMaxScaler()

X['Mileage'] = scaler.fit_transform(X[['Mileage']])
```

```python
X
```

#### **Data Splitting**

```python
[1, 2, 3, 4, 5, 6]
rs=42

[4, 5, 3, 1, 2, 6]

rs=3
[5, 6, 4, 3, 1, 2]

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.8, shuffle=True, random_state=42)
```

```python
X_test.shape
```

## **Data Modeling**

```python
lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
y_test
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
sgd = SGDRegressor()

sgd.fit(X_train, y_train)
```

```python
y_pred = sgd.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
print(mean_squared_error(y_test, y_pred))
```

```python
lasso = Lasso()

lasso.fit(X_train, y_train)
```

```python
y_pred = lasso.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
```

```python
ridge = Ridge()

ridge.fit(X_train, y_train)
```

```python
y_pred = ridge.predict(X_test)
```

```python
print(mean_absolute_error(y_test, y_pred))
```
